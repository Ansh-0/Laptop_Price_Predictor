```python
cat_data = cat_data.drop(['Os', 'TypeName', 'Company'], axis=1)
```

```python
print(cat_data['Cpu_brand'].value_counts())
```

```python
cpu_dict = {
    'Other Intel Processor': 0,
    'Intel Core i3': 1,
    'AMD Processor': 2,
    'Intel Core i5': 3,
    'Intel Core i7': 4
}
gpu_dict = {
    'AMD': 0, 'Intel': 1, 'Nvidia': 2
}
```

```python
cat_data['Cpu_brand'] = cat_data['Cpu_brand'].map(cpu_dict)
cat_data['Gpu_brand'] = cat_data['Gpu_brand'].map(gpu_dict)
```

```python
cat_data
```

```python
num_data
```

```python
num_data['Weight'] = num_data['Weight'].round().astype('int')
```

```python
num_data['SSD'].value_counts()
```

```python
def categorize_Ppi(ppi):
    if 90 <= ppi < 143: return 0
    elif 143 <= ppi < 196: return 1
    elif 196 <= ppi < 249: return 2
    elif 249 <= ppi < 302: return 3
    else: return 4
```

```python
num_data['Ppi'] = num_data['Ppi'].apply(categorize_Ppi)
```

```python
num_data
```

```python
complete_df = pd.concat([cat_data, num_data], axis=1)
complete_df.head()
```

```python
x = complete_df.drop('Price', axis=1)
y = complete_df[['Price']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=24)
rdf = RandomForestRegressor()
```

```python
rdf.fit(x_train, y_train)
```

```python
rdf.score(x_train, y_train)
```

```python
rdf.score(x_test, y_test)
```

```python
pred = rdf.predict(x_test)
compare = y_test
compare['Prediction'] = pred
compare
```
